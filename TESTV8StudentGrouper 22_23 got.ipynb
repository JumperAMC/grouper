```python
import pandas as pd, numpy as np, math

#import data from the excel file and turns it into a dataframe
df = pd.read_excel('immersive23sepfiltered.xlsx')
#describes the data to give number of students - will need to do 'manual' counts for groups sorting sizes

## below Randomises the order of rows so that alphabetical names are reorgansied across rows and reindexed
dfrandomised = df.sample(frac=1)

#Sort as male and female so women are filtered in first to groups by cross section routine
dfrandomgendersort = dfrandomised.sort_values(by = 'Gender').reset_index(drop=True)

dfrandom = dfrandomgendersort

#Groups the data by course and gender into something we can work with
gendergroups = dfrandom.groupby(["PROGRAMME OF STUDY","Gender"]).groups#,"Gender"#

#gives a list of names to use as keys in the gendergroups dictionary
gender_keys_list = list(gendergroups)

#extract the variables for the numpy arrays I will need:
total_female_students = dfrandom['Gender'].value_counts().f
total_male_students = dfrandom['Gender'].value_counts().m
total_students = len(dfrandom.index)
total_programmes = len(dfrandom.groupby("PROGRAMME OF STUDY").groups)
total_gender_programmes = len(gendergroups)
#THIS WORKS!!!! DO NOT TOUCH!!!!
#i in and of itself is the key valueistelf - i.e the course and gender filter.
#this line prints the total number of students from counting the index
print(dfrandom['PROGRAMME OF STUDY'].describe())
print("\nTotal number of students: ", total_students)
print('Total number of programmes : ',total_programmes)
print("Total gender and programme groups: ",total_gender_programmes,'\n')
print("Total female students: ",total_female_students,'\n')
print("Total male students: ",total_male_students,'\n')
print('--------------------')
```
